```
from pyspark.sql.functions import *

file_type = "csv"

# CSV options
inferSchema = "true"
firstRowIsHeader = "true"
delimiter = ","
```

```
def processDF(fileLocation):
    return(spark.read.csv(fileLocation, inferSchema=inferSchema, header=firstRowIsHeader, sep=delimiter))
    

athleteDF=processDF("/FileStore/tables/new_athlete_data.csv")
display(athleteDF)
winterAthleteDF=processDF("/FileStore/tables/athletes_winter_games.csv")
display(winterAthleteDF)
```

```
gamesDF=processDF("/FileStore/tables/games.csv")
display(gamesDF)
```

```
citiesDF=processDF("/FileStore/tables/cities.csv")
display(citiesDF)
```

```
sportsDF=processDF("/FileStore/tables/sports.csv")
display(sportsDF)
```

```
teamsDF=processDF("/FileStore/tables/teams.csv")
display(teamsDF)
```

```
olympicsDF=athleteDF.join(winterAthleteDF, athleteDF['id'] == winterAthleteDF['athlete_id'])\
      .join(gamesDF, gamesDF['game_id'] == winterAthleteDF['game_id'])\
      .join(citiesDF, citiesDF['city_id'] == gamesDF['city_id'])\
      .join(sportsDF, sportsDF['sport_id'] ==  winterAthleteDF['sport_id'])\
      .join(teamsDF, teamsDF['team_id'] == winterAthleteDF['team_id'])


display(olympicsDF)
```

```
medalistDF=olympicsDF.select(col("name"), col("medal"), col("city"), col("noc"), col("year"))
display(medalistDF)
```
