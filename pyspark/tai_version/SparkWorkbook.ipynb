```
def closeAdjClose(fileIs):
  splitData = fileIs.map(lambda x : x.split(splitIt))
  RDD = splitData.map(lambda x: (float(x[2]), float(x[6]))).sortByKey(True,1)
  return RDD
```

```
def maxHighDate(fileIs):
  splitData = fileIs.map(lambda x : x.split(splitIt))
  RDD1 = splitData.map(lambda x: (x[0], float(x[2]))).sortByKey(True,1)
  return RDD1
```

```
def minLowDate(fileIs):
  splitData = fileIs.map(lambda x : x.split(splitIt))
  RDD1 = splitData.map(lambda x: (x[0], float(x[3]))).sortByKey(True,1)
  return RDD1
```
