```
fileLocation = "/FileStore/tables/yahoo_daily.csv"
fileType = "csv"
inferSchema="true"
firstRowIsHeader="true"
delimiter=","

yahooDF=spark.read.format(fileType)\
    .option("inferSchema", inferSchema)\
    .option("header", firstRowIsHeader)\
    .option("sep", delimiter)\
    .load(fileLocation)
  
```

```
display(yahooDF)
```

```
yahooDF.show(100)
```

```
from pyspark.sql.functions import *


newYahooDF=yahooDF.select(col("date").alias("YahooDate"), col("open").alias("Opening Yahoo"),col("Close").alias("Closing Yahoo"), col("volume").alias("Yahoo Volume"))

print("\n\nOpening and Closing Yahoo Trades")
display(newYahooDF)
```

```
newYahooDF.printSchema()
```

```
newYahooDF=newYahooDF.withColumn("Year", year(newYahooDF.YahooDate))

newYahooDF.show(8000)
display(newYahooDF)
```
